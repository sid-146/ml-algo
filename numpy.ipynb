```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
```

```python
def func(X):
    return X * 2


transformed_func = FunctionTransformer(func)

pipeline = Pipeline([("custom_step", transformed_func)])


print(pipeline.fit_transform(X))
```

```python
import numpy as np
```

```python
X = np.array([[1, 2], [3, 4], [8, 5]])
y = np.array([[3], [7], [12]])
```

```python
X, y
```

```python
x_t = X.T
x_t
```

```python
A = np.dot(x_t, X)
A
```

```python
inverse = np.linalg.inv(A)
inverse
```

```python
result = np.dot(np.dot(inverse, x_t), y)
result
```
