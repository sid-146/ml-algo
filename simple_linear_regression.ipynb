## Through OLS

### Formula

$$\^{\beta }=(\mathbf{X}^{T}\mathbf{X})^{-1}\mathbf{X}^{T}\mathbf{y}$$

```python
import numpy as np
```

```python
X_train_new = np.c_[np.ones(len(X_train)), X_train]
x_t = X_train_new.T
x_inv = np.linalg.inv(x_t @ X_train_new)
betas = x_inv @ x_t @ y_train
```

```python
y_pred = betas[0] * X_test + betas[0]
```
